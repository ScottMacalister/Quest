<h1 style="color: #FF8C00;">Exercise 2: Daily Percentage Returns</h1>

---

You are required to calculate the daily percentage returns for each financial asset. Utilize this data to accomplish the following:
- Calculate the correlation matrix for the five assets.
- Create a scatter plot comparing the returns of two specific assets.

<h2 style="color: #FF6347;">Daily Returns Calculation.</h2>
<h3 style="color: #FF6347;">-- Insert your code here --</h3>

```python
returns = df.set_index('date').pct_change(fill_method=None) * 100
returns
```

<h2 style="color: #FF6347;">Correlation Matrix Calculation.</h2>
<h3 style="color: #FF6347;">-- Insert your code here --</h3>

```python
correlation_matrix = returns.corr()
print("Correlation Matrix:\n", correlation_matrix)
```

<h2 style="color: #FF6347;">Scatter Plot between the Returns of Two Assets.</h2>
<h3 style="color: #FF6347;">-- Insert your code here --</h3>

```python
plt.figure(figsize=(8, 6))
plt.scatter(returns['Asset1'], returns['Asset5'], alpha=0.7)
plt.title('Scatter Plot of Asset 1 vs Asset 5 Returns')
plt.xlabel('Asset 1 Returns (%)')
plt.ylabel('Asset 5 Returns (%)')
plt.grid(True)
plt.show()
```

<h1 style="color: #FF8C00;">Exercise 3: Portfolio Analysis</h1>

This third exercise focuses on calculating and analyzing the performance of a portfolio. Begin by loading the file named `portfolio_weights.csv`, which contains the daily weights of each asset in the portfolio.

To complete the exercise, you will need to perform the following tasks:
- Create an area chart of the asset weights.

- Determine the annualized volatility of the portfolio (using an annualization factor of 261 days).

```python
df2 = pd.read_csv('portfolio_weights.csv')
```

```python
df2.head(5)
```

```python
df2.tail(5)
```

<h2 style="color: #FF6347;">Area Chart of Asset Weights.</h2>
<h3 style="color: #FF6347;">-- Insert your code here --</h3>

```python
df2.plot(kind='area', stacked=True, figsize=(10, 6), alpha=0.7)
plt.title('Area Chart of Asset Weights Over Time')
plt.xlabel('Date')
plt.ylabel('Asset Weights')
plt.legend(title='Assets')
plt.show()
```

<h2 style="color: #FF6347;">Chart of Historical Cumulative Returns of the Portfolio.</h2>
<h3 style="color: #FF6347;">-- Insert your code here --</h3>

<h2 style="color: #FF6347;">Annualized Return.</h2>
<h3 style="color: #FF6347;">-- Insert your code here --</h3>

```python
cumulative_returns = df2.cumsum()

cumulative_returns.plot(figsize=(10, 6))
plt.title('Historical Daily Cumulative Returns of Each Asset')
plt.xlabel('Date')
plt.ylabel('Cumulative Return')
plt.legend(title='Assets')
plt.show()
```

<h2 style="color: #FF6347;">Annualized Volatility.</h2>
<h3 style="color: #FF6347;">-- Insert your code here --</h3>

```python
df2['date'] = pd.to_datetime(df2['date'], errors='coerce')
df2.set_index('date', inplace=True)

df2 = df2.apply(pd.to_numeric, errors='coerce')

cumulative_returns = df2.cumsum().iloc[-1]

portfolio_cumulative_return = cumulative_returns.mean()

days = (df2.index[-1] - df2.index[0]).days
years = days / 365

annualized_return = (1 + portfolio_cumulative_return) ** (1 / years) - 1

print(f"Annualized Return of the Portfolio: {annualized_return:.2%}")
```

<h2 style="color: #FF6347;">Area Chart of Asset Weights Grouped by Family.</h2>
<h3 style="color: #FF6347;">-- Insert your code here --</h3>

```python
df3 = pd.read_csv('asset_information_data.csv')
```

```python
df3
```

```python
df2_with_family = df2.copy()
df2_with_family.columns = df3['Name']
df2_with_family = df2_with_family.T
df2_with_family = df2_with_family.merge(df3, left_index=True, right_on='Name')
df2_with_family.set_index('Name', inplace=True)

category_weights = df2_with_family.groupby('Family').sum().T

category_weights.plot(kind='area', stacked=True, figsize=(10, 6), alpha=0.7)
plt.title('Area Chart of Asset Weights by Category')
plt.xlabel('Date')
plt.ylabel('Category Weights')
plt.legend(title='Asset Categories')
plt.show()
```
